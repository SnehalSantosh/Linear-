```python
train['vs'].fillna((train['vs'].mean()))
train['am'].fillna((train['am'].mean()))
train['gear'].fillna((train['gear'].mean()))
train['carb'].fillna((train['carb'].mean()))

train
```

```python
mylist = list(train.select_dtypes(include=['object']).columns)
mylist
```

```python
dummies = pd.get_dummies(train[mylist], prefix= mylist)

dummies.isnull().sum()
```

```python
train.drop(mylist, axis=1, inplace = True)

X = pd.concat([train,dummies], axis =1 )

train.isnull().sum()
```

```python
X = train.loc[:,['mpg','cyl','disp','hp','drat','qsec','vs','am','gear','carb']]

X
```

```python
x_train, x_cv, y_train, y_cv = train_test_split(X,train.wt, test_size = 0.3)

y_cv
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()

lreg.fit(x_train,y_train)


pred = lreg.predict(x_cv)
```

```python
coeff = DataFrame(x_train.columns)

coeff['Coefficient Estimate'] = Series(lreg.coef_)


coeff
```

```python
mse = np.mean((pred - y_cv)**2)

mse
```

```python
lreg.score(x_cv,y_cv)

```

```python
x_plot = plt.scatter(pred, (pred - y_cv), c='b')

plt.hlines(y=0, xmin= -1000, xmax=5000)

plt.title('Residual plot')
```
